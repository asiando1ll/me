震源位置：岩手県山田町の三陸沖の太平洋プレート
地震発生日：17時54分ごろ
マグニチュード：6.4
震源の深さ：10キロ（かなり浅い）
深さ10キロと浅いため、エネルギーが地表に伝わりやすく、沿岸部で強い揺れを感じた可能性は考えられる。
周辺のマップを見ると、過去の地震が非常に密集しており、プレート境界型地震が頻発するエリアだとわかる。
発生場所と深さから、太平洋プレートが沈み込む境界付近での逆断層型地震の可能性が高い。
このタイプの地震は津波を伴うことが多いらしい。
震源が浅く、規模が6.4と比較的大きいことから、津波注意報レベルの発表がありえる規模。
